```python
import folium
import csv

# Bahrain coordinates
bahrain_coord = [26.0667, 50.5577]

# Create a map centered on Bahrain
map_bahrain = folium.Map(location=bahrain_coord, zoom_start=10)

# Add a marker at the center of Bahrain
folium.Marker(location=bahrain_coord, popup='Bahrain').add_to(map_bahrain)

# import csv file
with open('Bahrain-Data-One.csv', encoding='utf-8', errors='ignore') as f:
    reader = csv.reader(f)
    next(reader) # skip headers
    data = list(reader)
    
# add markers for each location in the csv file
for row in data:
    folium.Marker(location=[float(row[9]), float(row[8])], popup=row[1]).add_to(map_bahrain)

# Display the map
map_bahrain

```

```python
import folium
import csv


# Set the location and zoom level for the map
map_center = [26.0667, 50.5577]
zoom_level = 10

# Create a Folium map centered on Bahrain
m = folium.Map(location=map_center, zoom_start=zoom_level)

# add a marker for the center of Bahrain
folium.Marker(location=map_center, popup='Bahrain').add_to(m)

# import the Bahrain petrol station .csv file
with open('Bahrain-Petrol-Stations.csv', encoding='utf-8', errors='ignore') as f:
    reader = csv.reader(f)
    next(reader) # skip headers
    data = list(reader)
    
# add markers for each location in the csv file
for row in data:
    folium.Marker(location=[float(row[9]), float(row[8])], popup=row[1], icon=folium.Icon(color='red', icon='info-sign')).add_to(m)

# import the Bahrain Data One .csv file
with open ('Bahrain-Data-One.csv', encoding='utf-8', errors='ignore') as f2:
    reader = csv.reader(f2)
    next(reader) # skip headers
    dataTwo = list(reader)
    
# add markers for each location in the csv file
for row in dataTwo:
    folium.Marker(location=[float(row[9]), float(row[8])], popup=row[1]).add_to(m)
    folium.Circle(location=[float(row[9]), float(row[8])], radius=500, fill_color='#3186cc', fill_opacity=0.2).add_to(m)
    


# Add Google Earth tiles to the map
folium.TileLayer(
    tiles='https://www.google.com/maps/vt?lyrs=s&x={x}&y={y}&z={z}',
    attr='Google',
    name='Google Earth',
    overlay=True,
).add_to(m)

# Display the map
m
```
